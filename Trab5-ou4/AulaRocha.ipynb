```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
class Xtreme_NN(st.rv_continuous):

    def __init__(self, N):
        self.N = N
        self.X = st.norm(25, 1)             # change this to change origin
        st.rv_continuous.__init__(self)

    def _cdf(self, x):
        return (self.X.cdf(x))**self.N
```

```python
x   = np.linspace(22, 29, 200)

f1  = plt.figure(1, figsize=(12, 4), clear=True)

for N in [1, 2, 8, 32, 128]:
    XN  = Xtreme_NN(N)

    f11 = plt.subplot(1,2,1)
    f1b = plt.plot(x, XN.pdf(x), label=N)

    f12 = plt.subplot(1,2,2)
    f1b = plt.plot(x, XN.cdf(x), label=N)

f11 = plt.subplot(1,2,1)
plt.title('Probability densities')
plt.grid(True)

f12 = plt.subplot(1,2,2)
plt.title('Cumulative distributions')
plt.grid(True)

plt.legend()
```
